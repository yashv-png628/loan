```python
loan_dataset=loan_dataset.dropna()
loan_dataset.info()
```

```python
loan_dataset.info()
```

```python
#data preprocessing
```

```python
loan_dataset['Dependents'].fillna(1,inplace=True)
loan_dataset.info()
```

```python
loan_dataset['LoanAmount'].fillna(loan_dataset.LoanAmount.mean(),inplace=True)
loan_dataset.info()
```

```python
Value_mapping={'Yes':1,'No':0}
loan_dataset['Married_Section']=loan_dataset['Married'].map(Value_mapping)
loan_dataset.head()
```

```python
Value_mapping1={'Male':1,'Female':0}
loan_dataset['Gender_Section']=loan_dataset['Gender'].map(Value_mapping1)
loan_dataset.head()
```

```python
loan_dataset['Education'].unique()
```

```python
Value_mapping2={'Gradute':1,'not Gradute':0}
loan_dataset['Edu_Section']=loan_dataset['Education'].map(Value_mapping2)
loan_dataset.head()
```

```python
loan_dataset.info()
```

```python
loan_dataset['Married_Section'].fillna(loan_dataset.Married_Section.mean(),inplace=True)
loan_dataset['Gender_Section'].fillna(loan_dataset.Married_Section.mean(),inplace=True)
loan_dataset['Loan_Amount_Term'].fillna(loan_dataset.Married_Section.mean(),inplace=True)
loan_dataset['Credit_History'].fillna(loan_dataset.Married_Section.mean(),inplace=True)
loan_dataset.info()
```

```python
Value_mapping3={'Yes':1,'No':0}
loan_dataset['Employed_Section']=loan_dataset['Self_Employed'].map(Value_mapping3)
loan_dataset.head()
```

```python
loan_dataset.info()
```

```python
loan_dataset['Employed_Section'].fillna(loan_dataset.Employed_Section.mean(),inplace=True)
loan_dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make=LabelEncoder()
loan_dataset['Property_Section']=lb_make.fit_transform(loan_dataset['Property_Area'])
loan_dataset.head()
```

```python
Value_mapping4={'Y':1,'N':0}
loan_dataset['Loan_Section']=loan_dataset['Loan_Status'].map(Value_mapping4)
loan_dataset.head()
```
